```python
import pandas as pd
import matplotlib.pyplot as plt

df_participe = pd.read_csv('participe.csv')
df_est_comp = pd.read_csv('est_composee.csv')
df_constitue = pd.read_csv('constitue.csv')

df_menu = pd.merge(df_participe, df_est_comp, on='idReunion')
df_menu = pd.merge(df_menu, df_constitue, on='idPlat')

ingredient_counts = df_menu['nomA'].value_counts()

plt.figure(figsize=(10, 8))


explode = [0.1 if nom == 'Fromage à raclette' else 0 for nom in ingredient_counts.index]

ingredient_counts.plot(
    kind='pie', 
    autopct='%1.1f%%', 
    startangle=140, 
    explode=explode, 
    cmap='Set3',
    textprops={'fontsize': 11, 'fontweight': 'bold'}
)

plt.title("Pourcentage d'aliment le plus mangé en réunion", fontsize=15, fontweight='bold')
plt.ylabel("")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df_territoire = pd.read_csv('TERRITOIRE.csv')
df_orga = pd.read_csv('ORGANISATION.csv')
df_club = pd.read_csv('CLUB.csv')
df_reunion = pd.read_csv('REUNION.csv')
df_lieu = pd.read_csv('LIEU.csv')
df_est_comp = pd.read_csv('est_composee.csv')
df_aliments = pd.read_csv('constitue.csv')



df_ordres = pd.DataFrame({'numOrganisation': range(1, 14), 'idTerritoire': range(1, 14)})


df_club_renomme = df_club.rename(columns={'numOrganisation_1': 'numOrganisation', 'numOrganisation': 'idTerritoire'})


mapping_orga_terr = pd.concat([df_ordres, df_club_renomme])
mapping_orga_terr = pd.merge(mapping_orga_terr, df_territoire, on='idTerritoire')


df_geo_reunion = pd.merge(df_reunion, df_lieu, on='idLieu')
df_geo_reunion = pd.merge(df_geo_reunion, mapping_orga_terr, on='numOrganisation')
df_geo_plats = pd.merge(df_geo_reunion, df_est_comp, on='idReunion')
df_geo_menu = pd.merge(df_geo_plats, df_aliments, on='idPlat')


df_fromage = df_geo_menu[df_geo_menu['nomA'] == 'Fromage à raclette']


classement_territoires = df_fromage.groupby('nomTerritoire')['nomA'].count().sort_values(ascending=False)


plt.figure(figsize=(12, 6))
classement_territoires.plot(kind='bar', color='orange', edgecolor='black')

plt.title("Classement des territoires mangeant le plus de Fromage à raclette")
plt.xlabel("Territoire")
plt.ylabel("Nombre de portions servies")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_tenrac = pd.read_csv('TENRAC.csv')
df_participe = pd.read_csv('participe.csv')
df_est_comp = pd.read_csv('est_composee.csv')
df_aliments = pd.read_csv('constitue.csv')

df_reunions = pd.merge(df_tenrac[['codeTenrac', 'nomGrade']], df_participe, on='codeTenrac')
df_plats = pd.merge(df_reunions, df_est_comp, on='idReunion')
df_menu_final = pd.merge(df_plats, df_aliments, on='idPlat')

comptage = df_menu_final.groupby(['nomGrade', 'nomA']).size().reset_index(name='Nombre')
comptage_trie = comptage.sort_values(['nomGrade', 'Nombre'], ascending=[True, False])
top_aliment_par_grade = comptage_trie.drop_duplicates(subset=['nomGrade'])

top_aliment_par_grade = top_aliment_par_grade.sort_values('Nombre', ascending=True)

plt.figure(figsize=(12, 8))
plt.style.use('seaborn-v0_8-whitegrid')
barres = plt.barh(top_aliment_par_grade['nomGrade'], top_aliment_par_grade['Nombre'], color='steelblue', edgecolor='black')

for barre in barres:
    label = "{:.0f}".format(barre.get_width())
    plt.text(barre.get_width() + 100, barre.get_y() + barre.get_height()/2, label, 
             va='center', ha='left', fontsize=11, fontweight='bold', color='black')

plt.title("Palmarès des Consommations par Grade : Le Fromage à Raclette en Tête", fontsize=15, fontweight='bold')
plt.xlabel("Nombre de fois consommé", fontsize=12)
plt.ylabel("Grade des Membres", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd


df_tenrac = pd.read_csv('TENRAC.csv')
df_participe = pd.read_csv('participe.csv')
df_reunion = pd.read_csv('REUNION.csv')
df_est_comp = pd.read_csv('est_composee.csv')
df_constitue = pd.read_csv('constitue.csv')


df_fromage = df_constitue[df_constitue['nomA'] == 'Fromage à raclette']


df_reunion_fromage = pd.merge(df_est_comp, df_fromage, on='idPlat')
portions_par_reunion = df_reunion_fromage.groupby('idReunion').size().reset_index(name='portions_servies')


df_toutes_reunions = pd.merge(df_reunion[['idReunion']], portions_par_reunion, on='idReunion', how='left')
df_toutes_reunions['portions_servies'] = df_toutes_reunions['portions_servies'].fillna(0)


df_repas_membres = pd.merge(df_participe, df_toutes_reunions, on='idReunion')
df_repas_complet = pd.merge(df_repas_membres, df_tenrac[['codeTenrac', 'nomGrade']], on='codeTenrac')


stats_par_repas = df_repas_complet.groupby('nomGrade')['portions_servies'].agg(
    Médiane='median',
    Moyenne='mean',
    Écart_Type='std',
    Maximum='max'
).round(2)

print("\n")
print("  STATISTIQUES DE PLATS À RACLETTE PAR RÉUNION ET PAR GRADE")
print("\n")
print(stats_par_repas)
```
